```python
import requests as rq
import pandas as pd
import re
from bs4 import BeautifulSoup as bs
```

```python
companies = []
for i in range(1, 12):
    url = 'https://companies.rbc.ru/okved/10.61/{}/'.format(str(i))
    resp = rq.get(url)
    page = bs(resp.text, 'html.parser')
    for i in page.find_all('div', {'class' : 'company-card info-card'}):
        name = i.find_all('a')[0].find_all('span')[0].text
        work = i.find_all('span', {'class' : 'company-status-badge company-status-badge--green'})
        revenue = i.find_all('div', {'class': 'company-card__block'})[2].find_all('p')[0].text.split(':')[1]. \
              replace('\xa0', '').replace(',0₽', '')
        address_prep = i.find_all('p')
        companies.append({'name': name, 'is_working': work[0].text if work else None, 'revenue' : revenue, 
                          'address': address_prep[2].text if address_prep[1].text.startswith('Ген') else address_prep[1].text})
```

```python
pd.DataFrame(companies).to_excel('companies.xlsx')
```
