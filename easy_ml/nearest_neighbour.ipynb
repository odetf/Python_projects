# Предобработка данных

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
y = df['Survived'] #целевая переменная
data = df.drop(columns=['Survived']) #признаки
data.head(10)
```

```python
data = data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
data.head(10)
```

```python
data.info()
```

```python
#заполняем пропуски для age средним значением
data['Age'] = data['Age'].fillna(np.mean(data['Age']))
```

```python
#рассмотрим теперь embarked
data['Embarked'].unique()
```

```python
data['Embarked'].value_counts()
```

```python
data['Embarked'] = data['Embarked'].fillna('S')
```

```python
data.head(10)
```

```python
#Кодируем пол
sex_dict = {'female': 0, 'male':1}
for i in range(len(data)):
    data['Sex'][i] = sex_dict[data['Sex'][i]]
```

```python
#закодируем класс обслуживания через sklearn
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(data['Embarked']) #кодируем
data['Embarked'] = le.transform(data['Embarked']) #трансформируем

data.head()
```

```python
data.corr()
```

# Обучение модели

```python
from sklearn.metrics import accuracy_score #доля верных ответов алгоритма
from sklearn.metrics import roc_auc_score #площадь под кривой ошибок
from sklearn.metrics import precision_score #точность
from sklearn.metrics import recall_score #полнота
```

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
```

KNN

```python
knn = KNeighborsClassifier(n_neighbors = 5)

knn.fit(data, y)
knn_preds = knn.predict(data) #присвоить лейбл каждому из классов
knn_preds_proba = knn.predict_proba(data)[:,1] #оценка вероятностей для тестовых данных

print('accuracy', accuracy_score(y, knn_preds))
print('roc_auc', roc_auc_score(y, knn_preds_proba))
```
