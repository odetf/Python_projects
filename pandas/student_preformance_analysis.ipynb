```python
st_perf = st_perf.rename(columns = {'parental level of education' : 'parental_level_of_education',
                                   'test preparation course': 'test_reparation_course',
                                    'math score': 'math_score',
                                    'reading score' : 'reading_score',
                                    'writing score' : 'writing_score'})
```

```python
st_perf.query('gender == "female" & writing_score > 78')
```

```python
writing_score_query = 78
st_perf.query('writing_score > @writing_score_query')
```

```python
#The same result but with query
parent_high = ["bachelor's degree", "master's degree"]
st_perf.query('parental_level_of_education == @parent_high')
```
