```python
import random
```

```python
random.seed("АДВИ_2023")
N = 1000

index = [x for x in range(1, N+1)]
random.shuffle(index)

groups = random.choices(["группа_1", "группа_2", "группа_3", "группа_4", "группа_5"], k = 1000)

uniform = [random.randint(50, 100) for i in range(N)]

gauss_1 = [random.gauss(mu = 0, sigma=1) for i in range(N)]
gauss_2 = [random.gauss(mu = 0, sigma=12) for i in range(N)]
gauss_3 = [random.gauss(mu = 50, sigma=9) for i in range(N)]
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'index':index, 'groups':groups, 'uniform':uniform, 'gauss_1':gauss_1, 'gauss_2':gauss_2, 'gauss_3':gauss_3})
df = df.set_index('index')
df.head()
```

```python
df.dtypes
```

```python
for i in df['gauss_1']:
    if df['gauss_1'][df['gauss_1'] == i].index % 121 == 0:
        df['gauss_1'][df['gauss_1'] == i] = None
```

```python
for i in df['gauss_2']:
    if i % 1 > 0.95:
        df['gauss_2'][df['gauss_2'] == i] = None
```

```python
df.gauss_2.mean()
```

```python
df['gauss_2'].fillna(df.gauss_2.mean(), inplace=True)
df['gauss_2']
```
